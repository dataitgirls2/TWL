```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_edge(1, 2)
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
g = nx.Graph()
g.add_node('apple')
g.add_nodes_from(['banana', 'kiwi', 'mango'])
```

```python
g.nodes()
```

```python
g = nx.Graph()
g.add_edge('apple', 'banana')
g.add_edges_from([('apple', 'mango'), ('apple', 'kiwi')])
```

```python
g.edges()
```

```python
g.nodes()
```

```python
g.node['kiwi']['kind'] = 'fruit'
g.node['kiwi']
```

```python
g.nodes(data=True)
```

```python
g.add_node('kiwi', kind='fruit')
g.add_nodes_from(['banana', 'apple'], kind='fruit')
g.node['banana']
```

```python
g.edges(data=True)
```

```python
nx.draw_spring(g, with_labels=True)
```

```python
g.edges(data=True)
```

```python
relations = [('apple', 'mango'), ('apple', 'kiwi')]
g.add_edge('apple', 'mango', wieght=2.5)
g.add_edges_from(relations, color='blue')
g['apple']['mango']['weight'] = 5
```

```python
nx.draw(g, with_labels=True)
```

```python
G = nx.MultiGraph()
G.add_nodes_from(['A','B','C','D'])
G.add_edges_from([('A','B'),('A','B'),('A','D'),('B','D'), ('B','C'), ('B','C'),('C','D')])
nx.draw(G, with_labels=True)
plt.show()
```

```python
G1 = nx.Graph()
G1.add_edges_from([(1,2), (2,3), (3,4)])
nx.draw(G1, with_labels=True)
plt.show()
```

```python
G2 = nx.DiGraph([(1,2), (2,3), (3,4)])
nx.draw(G2, with_labels=True)
plt.show()
```
